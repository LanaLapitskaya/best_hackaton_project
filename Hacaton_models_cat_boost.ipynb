```python
import catboost
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('prep.csv')
```

```python
#data.iloc[:, :50].head().T
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop('cost', axis=1), 
                                                    data.cost.str.replace(' ', '').astype('int32'),
                                                    test_size=0.2)
```

```python
ctbst = catboost.CatBoostRegressor()
```

```python
ctbst.fit(X_train.iloc[:, :49], y_train)
```

```python
r2_score(y_test, ctbst.predict(X_test.iloc[:, :49]))
```
